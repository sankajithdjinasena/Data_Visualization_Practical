# Lab: Data Visualization & Regression Mastery

## 1. Environment Setup & Data Loading

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load the tips dataset
tips_df = sns.load_dataset('tips')

# Display the first 5 rows of the DataFrame
display(tips_df.head())
```

## Visualization Tasks

```python
# Task 1: The Baseline Regression
plt.figure(figsize=(8, 6))
sns.regplot(x='total_bill', y='tip', data=tips_df, line_kws={'color': 'red'})
plt.title('Total Bill vs. Tip (Baseline Regression)')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', col='time', row='smoker', data=tips_df, aspect=0.6)
plt.suptitle('Total Bill vs. Tip by Time of Day and Smoking Status', y=1.02)
plt.show()
```

```python
sns.pairplot(tips_df, hue='day', kind='reg')
plt.suptitle('Pairplot of Tipping Data by Day with Regression Lines', y=1.02)
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

lunch_df = tips_df[tips_df['time'] == 'Lunch']
dinner_df = tips_df[tips_df['time'] == 'Dinner']

sns.regplot(x='total_bill', y='tip', data=lunch_df, ax=axes[0], line_kws={'color': 'green'})
axes[0].set_title('Total Bill vs. Tip for Lunch')
axes[0].set_xlabel('Total Bill ($)')
axes[0].set_ylabel('Tip ($)')

sns.regplot(x='total_bill', y='tip', data=dinner_df, ax=axes[1], line_kws={'color': 'blue'})
axes[1].set_title('Total Bill vs. Tip for Dinner')
axes[1].set_xlabel('Total Bill ($)')

plt.suptitle('Total Bill vs. Tip: Lunch vs. Dinner', y=1.05)
plt.tight_layout()
plt.show()
```
