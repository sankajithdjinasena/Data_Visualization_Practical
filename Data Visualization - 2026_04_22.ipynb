```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def add_labels(x,y):
  for i in range(len(x)):
    # plt.text(i,y[i],x[i])
    # plt.text(i,y[i],x[i], ha="center")
    # plt.text(i,y[i],x[i], ha="center", bbox = dict(facecolor="red", alpha=0.8))
    # plt.text(i,y[i]//2,y[i], ha="center")
    plt.text(i,y[i]//2,y[i], ha="center",bbox = dict(facecolor="white", alpha=0.5))

```

```python
x = ["FAGS","FAPS","FOC","FGS","FSS","FMS"]
y = [330,350,330,400,550,560]

plt.bar(x,y)
add_labels(x,y)

plt.title("Faculty Admission")
plt.xlabel("Faculty")
plt.ylabel("Number of Admissions")

plt.show()
```

```python
xl = np.linspace(0,5,100)
yl = np.cos(2*np.pi*xl) * np.exp(-xl)

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(12,5))

ax1.plot(xl,yl, color="red")
ax1.set_title("Plot without grid")
ax1.grid(False)

ax2.plot(xl,yl)
ax2.set_title("Plot with grid")
ax2.grid(True)

plt.show()
```

```python
fig, ax = plt.subplots(2,2,figsize=(12,10))
fig.suptitle("Plots")
x = np.linspace(0,10,1000)
ax[0,0].plot(x, np.sin(x), 'r-', label='sin(x)')
ax[0,1].plot(x, np.cos(x), 'g--', label='cos(x)')
ax[1,0].plot(x, np.tan(x), 'y-', label='tan(x)')
ax[1,1].plot(x, np.sinc(x), 'c.-', label='sinc(x)')

for axes in ax.flat:
  axes.legend()
  axes.grid(True)

plt.tight_layout()
plt.show()
```
